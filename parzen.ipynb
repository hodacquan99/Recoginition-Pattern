```
import numpy as np
import math
#ham tinh khoang cach
def dis_cal(X,Y):
  return np.linalg.norm(X-Y)
#Tinh Hypercube
def hypercube(X,Y,h):
  if (dis_cal(X,Y) <= h/2):
    return 1
  else:
    return 0
#Tinh phi
def p_cal(X , data ,h ):
  count = 0
  for x in data :
    count+=hypercube(X,x,h)
  return count
#Tinh Parzen Window
def PWindow(X,Data,h):
  k=0
  n = Data.shape[0]
  d = Data.shape[1]
  for x in Data:
    k+=hypercube(X,x,h)
    
  return k/n/h**d
```
